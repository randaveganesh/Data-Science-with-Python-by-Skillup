```python
auto_dataset = pd.read_csv(r"D:\Dataset1\auto_v1.csv")
```

```python
def origin(num):
    if num ==1:
        return "USA"
    elif num == 2:
        return "Europe"
    else:
        return "Asia"
    
auto_dataset["origin"] = auto_dataset["origin"].apply(origin)
```
